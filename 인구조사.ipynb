### '종로구'의 인구구조를, 0세~100세까지 나이대별로 몇 명이 있는지 살펴보기

```python
종로구
```

```python
search = '종로구'
종로구 = df.loc[df['행정구역'].str.contains(search),]
종로구 = 종로구.iloc[0, 3:]

x=np.arange(0, 100+1)
y = 종로구.values

plt.plot(x, y)
plt.title(search)
plt.xlabel('나이')
plt.ylabel('인구수')
plt.show()
```

### '종로구'의 인구구조를 만 0세, 15세, 25세, 35세, 45세로 나눠 인구구조 살펴보기

```python
search = '종로구'
종로구 = df.loc[df['행정구역'].str.contains(search), ]
종로구 = 종로구.iloc[0, 3:]
```

```python
종로구
```

```python
종로구 = 종로구[[0, 15, 25, 35, 45]]
```

```python
plt.pie(종로구, autopct='%.2f', labels=종로구.index, startangle=90, wedgeprops={'width' : 0.7})
plt.show()
```

```python
# 코드 하나로 짜기

search = '종로구'
종로구 = df.loc[df['행정구역'].str.contains(search), ]
종로구 = 종로구.iloc[0, 3:]

종로구 = 종로구[[0, 15, 25, 35, 45]]

plt.pie(종로구, autopct='%.2f', labels=종로구.index, startangle=90, wedgeprops={'width' : 0.7})
plt.show()
```
