### 컬럼명 변경하기

### 최근 3년간 CCTV 증가율을 계산하여, dataframe 에 '최근증가율' 컬럼을 추가하기

```python
CCTV_Seoul.head()
```

```python
CCTV_Seoul['최근증가율'] = ((CCTV_Seoul['2014년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2016년']) / CCTV_Seoul['2013년도 이전']) * 100
# 증가율 계산 : (2016 + 2015 + 2014) / 2013이전 * 100
```

```python
CCTV_Seoul['최근증가율']
```

```python
CCTV_Seoul.head()
```

### 최근 3년간 CCTV 증가율이 가장 높은 순으로 5개의 구 확인해보기

```python
CCTV_Seoul.sort_values('최근증가율', ascending=False).head(5)
```

### 서울시 인구데이터 중  첫번째 row('합계') 삭제하기

```python
pop_Seoul.head()
```

```python
pop_Seoul = pop_Seoul.iloc[ 1: , ]
```

```python
pop_Seoul.head()
```

### 서울시 인구데이터 중 NaN 값이 있는지 확인 후, NaN 값이 있다면 삭제하기

```python
pop_Seoul.isna().sum()
```

```python
pop_Seoul.dropna(inplace=True)
```

```python
pop_Seoul.isna().sum()
```

### '외국인비율' 과 '고령자비율' 컬럼 추가하기

```python
pop_Seoul.head()
```

```python
pop_Seoul['외국인비율'] = (pop_Seoul['외국인'] / pop_Seoul['인구수'])*100
```

```python
pop_Seoul['고령자비율'] = (pop_Seoul['고령자'] / pop_Seoul['인구수'])*100
```

```python
pop_Seoul.head()
```

### 외국인이 가장 많은 구와, 외국인비율이 가장 높은 구를 각각 5개 찾아보기

```python
pop_Seoul.sort_values('외국인', ascending=False).head()
```

```python
pop_Seoul.sort_values('외국인비율', ascending=False).head()
```

### 고령자가 가장 많은 구와, 고령자비율이 가장 높은 구를 각각 5개씩 찾아보기

```python
pop_Seoul.sort_values('고령자', ascending=False).head()
```

```python
pop_Seoul.sort_values('고령자비율', ascending=False).head()
```

### CCTV 데이터와 인구 데이터 합치고 분석해보기

```python
CCTV_Seoul.head()
```

```python
pop_Seoul.head()
```

```python
data1 = pd.merge(CCTV_Seoul, pop_Seoul, on='구별')
```

```python
data1
```

### 합쳐진 데이터에서 '2013년도 이전', '2014년', '2015년', '2016년' 컬럼 제외하기

```python
data2 = data1.drop(['2013년도 이전', '2014년', '2015년', '2016년'], axis=1)
```

```python
data2
```

### 데이터 인덱스를 '구별'로 바꾸기

```python
data2.set_index('구별', inplace=True)
```

```python
data2
```

### 'CCTV_result' 이름으로 CSV 파일로 저장하기

```python
data2.to_csv('CCTV_result')
```

### 상관분석하기 1. CCTV의 갯수와 고령자 비율의 상관관계 분석하기

```python
data2.head()
```

```python
plt.scatter(data=data2, x='고령자비율', y='소계')
plt.xlabel('고령자비율')
plt.ylabel('CCTV 갯수')
plt.show()
```

```python
sb.regplot(data=data2, x='고령자비율', y='소계')
plt.xlabel('고령자비율')
plt.ylabel('CCTV 갯수')
plt.show()

# 고령자비율과 CCTV 갯수 간에 거의 관계가 없거나, 아주 약간의 반비례 관계가 있다.
```

```python
data2[['소계', '고령자비율']].corr()
```

### 상관분석하기 2. CCTV의 갯수와 외국인 비율의 상관관계 분석하기

```python
sb.regplot(data=data2, x='외국인비율', y='소계')
plt.xlabel('외국인비율')
plt.ylabel('소계')
plt.show()

# 외국인비율과 CCTV 갯수 간에 관계가 없다
```

```python
data2[['소계', '외국인비율']].corr()
```

### 상관분석하기 3. CCTV의 갯수와 인구수의 상관관계 분석하기

```python
sb.regplot(data=data2, x='인구수', y='소계')
plt.xlabel('인구수')
plt.ylabel('소계')
plt.show()

# CCTV 갯수와 인구수 간에 약한 비례관계가 있다.
```

```python
data2[['소계', '인구수']].corr()
```
